```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Embedding,GRU,Dense
from keras.optimizers import Adam
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
dataset =pd.read_csv("reviews.txt",delimiter='\t',names=['Reviews','Sentiment'])
```

```python
dataset.head()
```

```python
X=dataset.drop(['Sentiment'],axis=1)
y=dataset.drop(['Reviews'],axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X.head()
```

```python
X_series = dataset.iloc[:,0].rename("Reviews")
#X_series['Reviews'] = X_series.Reviews.astype(str)
#X_series.head()
```

```python
num_words = 2000
```

```python
tokenizer = Tokenizer(num_words=num_words)
```

```python
%%time
tokenizer.fit_on_texts(X_series)
```

```python
if num_words is None:
    num_words = len(tokenizer.word_index)
```

```python
tokenizer.word_index
```

```python
x_train =X_train['Reviews']
x_test =X_test['Reviews']
Y_test = y_test['Sentiment']
Y_test.head()
```

```python
x_train_tokens = tokenizer.texts_to_sequences(x_train)
```

```python
type(x_train) 
```

```python
np.array(x_train_tokens[15])
```

```python
x_test_tokens = tokenizer.texts_to_sequences(x_test)
```

```python
np.array(x_test_tokens[15])
```

```python
num_tokens = [len(tokens) for tokens in x_train_tokens + x_test_tokens]
num_tokens = np.array(num_tokens)
```

```python
np.mean(num_tokens)
```

```python
np.max(num_tokens)
```

```python
pad ='pre'
max_tokens = 18
```

```python
x_train_pad = pad_sequences(x_train_tokens, maxlen=max_tokens,
                            padding=pad, truncating=pad)
```

```python
x_test_pad = pad_sequences(x_test_tokens, maxlen=max_tokens,
                            padding=pad, truncating=pad)
```

```python
x_train_pad.shape
```

```python
x_test_pad.shape
```

```python
np.array(x_train_tokens[1])
```

```python
x_train_pad[1]
```

## Building the RNN

```python
model = Sequential()
```

```python
model.add(Embedding(input_dim=num_words,
                   output_dim=8,
                   input_length=max_tokens,
                   name='layer_embedding'))
```

```python
model.add(GRU(units=16,return_sequences=True))
```

```python
model.add(GRU(units=8,return_sequences=True))
```

```python
model.add(GRU(units=4))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train_pad,y_train,epochs =100,batch_size=32)
```

```python
result = model.predict(x_test_pad)
```

```python
result=np.round(result,decimals=0)
```

```python
result=pd.DataFrame(result)
```

```python
result.head()
```

```python
cm = confusion_matrix(y_test, result)
```
