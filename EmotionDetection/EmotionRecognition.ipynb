```python
X_train = np.array([ np.fromstring(image, np.uint8, sep=" ").reshape((48,48)) for image in train.pixels])
```

```python
X_validation = np.array([ np.fromstring(image, np.uint8, sep=" ").reshape((48,48)) for image in validation.pixels])
```

```python
X_test = np.array([ np.fromstring(image, np.uint8, sep=" ").reshape((48,48)) for image in test.pixels])
```

```python
model = Sequential()
```

```python
model.add(Conv2D(64,(3,3),activation = 'relu',input_shape = (48,48,1), padding = "same"))
model.add(MaxPooling2D())
```

```python
model.add(Conv2D(64,(3,3),activation = 'relu'))
model.add(Conv2D(64,(3,3),activation = 'relu'))
model.add(AveragePooling2D())
```

```python
model.add(Conv2D(128,(3,3),activation = 'relu'))
model.add(Conv2D(128,(3,3),activation = 'relu'))
model.add(AveragePooling2D())
```

```python
model.add(Flatten())
```

```python
model.add(Dense(1024, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(1025, activation = 'relu'))
model.add(Dense(7, activation = 'softmax'))
```

```python
print(X_train[0].shape)
X_train = X_train.reshape((-1,48,48,1)).astype(np.float32)
print(X_train[0].shape)
X_train /= 255
print(X_train[0].shape)
X_validation = X_validation.reshape((-1,48,48,1)).astype(np.float32)
X_validation /= 255
X_test = X_test.reshape((-1,48,48,1)).astype(np.float32)
X_test /= 255
```

```python
model.compile(loss = "sparse_categorical_crossentropy", optimizer = tf.keras.optimizers.Adam(), metrics= ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,batch_size = 128, epochs = 30, validation_data = (X_validation,y_validation))
```

```python
def emotion_analysis(emotions):
    objects = ('angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral')
    y_pos = np.arange(len(objects))
    emotion_max = np.amax(emotions)
    emotion_max_index = np.where(emotions == emotion_max)
    result_emotion = objects[emotion_max_index[0][0]]
    plt.bar(y_pos, emotions, align='center', alpha=0.5)
    plt.xticks(y_pos, objects)
    plt.ylabel('percentage')
    plt.title('emotion')
    plt.show()
    return result_emotion
```

```python
def process():
    img = image.load_img("frame.jpg", grayscale=True, target_size=(48, 48))
 
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis = 0)
    x /= 255
    custom = saved_model.predict(x)
    emotion = emotion_analysis(custom[0])
    font = cv2.FONT_HERSHEY_SIMPLEX
    x = np.array(x, 'float32')
    x = x.reshape([48, 48]);
    #plt.gray()
    #plt.imshow(x)
    #plt.show()
    return emotion
```

```python
import cv2
from PIL import Image
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)
while (cap.isOpened()):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
        cv2.imwrite("frame.jpg", frame)
        process()
        
    cv2.imshow("Project", frame)
    
    if cv2.waitKey(1) == 13:
        break
        # 13 is the space key


cap.release()
cv2.destroyAllWindows()
```
